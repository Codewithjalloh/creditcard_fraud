```python
# 1. Plot a graph showing the Amount of transactions against Time.
plt.figure(figsize=(15, 7))
plt.scatter(data["Time"], data["Amount"], alpha=0.5, color='red')
plt.title("Amount vs Time")
plt.xlabel("Time (in seconds)")
plt.ylabel("Amount")
plt.show()
```

```python
# 2. Calculate the number of Fraudulent and Valid transactions in the dataset.
fraud_count = data[data["Class"] == 1].shape[0]
valid_count = data[data["Class"] == 0].shape[0]
print(f"Fraudulent Transactions: {fraud_count}")
print(f"Valid Transactions: {valid_count}")
```

```python
# 3. Determine the fraction of outliers (assuming the outliers are the fraudulent transactions).
outlier_fraction = fraud_count / (fraud_count + valid_count)
print(f"Outlier Fraction: {outlier_fraction:.5f}")
```

```python
# 4. Find the correlation between the features in the dataset.
correlation_matrix = data.corr()
print(correlation_matrix["Class"].sort_values(ascending=False))
```

```python
# 5. Visualise the correlation using a heatmap.
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```
